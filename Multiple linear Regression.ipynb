```python
def get_predictions(model,x):
    (n,p_minus_one)=x.shape
    p=p_minus_one+1
    new_x=np.ones(shape=(n,p))
    new_x[:,1:]=x
    return np.dot(new_x,model)
```

```python
test_model=np.array([1,2,3,4])
get_predictions(test_model,x_train)
```

```python
predictions_df['test prediction']=get_predictions(test_model,x)
predictions_df
```

```python
from numpy.linalg import inv
def get_best_model(model,x):
    
    n=x.shape[0]
    p_minus_one=x.shape[1]
    p=p_minus_one+1
    new_x=np.ones(shape=(n,p))
    new_x[:,1:]=x
    return np.dot(np.dot(inv(np.dot(new_x.T,new_x)),new_x.T),y)
```

```python
best_model=get_best_model(y,x)
predictions_df['best Prediction']=get_predictions(best_model,x)
predictions_df
```
